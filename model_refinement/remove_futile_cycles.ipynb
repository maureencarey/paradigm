```python
def met_ids_without_comp(model,met_id):
    
    if met_id in [met.id for met in model.metabolites]:
        for m in [model.metabolites.get_by_id(met_id)]:
            if m.id.endswith('_c') or m.id.endswith('_e') or m.id.endswith('_f') or m.id.endswith('_g')\
            or m.id.endswith('_h') or m.id.endswith('_i') or m.id.endswith('_l') or m.id.endswith('_m')\
            or m.id.endswith('_n') or m.id.endswith('_p') or m.id.endswith('_r') or m.id.endswith('_s')\
            or m.id.endswith('_u') or m.id.endswith('_v') or m.id.endswith('_x'):
                id_withou_c = m.id[:-2]
            elif m.id.endswith('_cx') or m.id.endswith('_um') or m.id.endswith('_im') or m.id.endswith('_ap')\
            or m.id.endswith('_fv'):
                id_withou_c = m.id[:-3]
            else:
                print('unknown compartment')
                print(m.id)
                id_withou_c = ''
    else:
        id_withou_c = ''
    return(id_withou_c)

def get_comp(model,met_id):
    
    if met_id in [met.id for met in model.metabolites]:
        for m in [model.metabolites.get_by_id(met_id)]:
            if m.id.endswith('_c') or m.id.endswith('_e') or m.id.endswith('_f') or m.id.endswith('_g')\
            or m.id.endswith('_h') or m.id.endswith('_i') or m.id.endswith('_l') or m.id.endswith('_m')\
            or m.id.endswith('_n') or m.id.endswith('_p') or m.id.endswith('_r') or m.id.endswith('_s')\
            or m.id.endswith('_u') or m.id.endswith('_v') or m.id.endswith('_x'):
                id_withou_c = m.id[-2:]
            elif m.id.endswith('_cx') or m.id.endswith('_um') or m.id.endswith('_im') or m.id.endswith('_ap')\
            or m.id.endswith('_fv'):
                id_withou_c = m.id[-3:]
            else:
                print('unknown compartment')
                print(m.id)
                id_withou_c = ''
    else:
        id_withou_c = ''
    return(id_withou_c)
```

```python
dict_m = dict()
for met in iPfal18.metabolites:
    l = list()
    met_no_c = met_ids_without_comp(iPfal18,met.id)
    print(met_no_c)
    for rxn in met.reactions:
        if met.id in [m.id for m in rxn.reactants]:
            for m in rxn.products:
                if met_no_c == met_ids_without_comp(iPfal18,m.id):
                    l.append(rxn)
        else: # met is a product
            for m in rxn.reactants:
                if met_no_c == met_ids_without_comp(iPfal18,m.id):
                    l.append(rxn)
    dict_m[met.id] = l
dict_m
```

```python
new_dict = dict()
for met_og,reactions in dict_m.items():
    if len(reactions)>1:
        comp_dict = dict()
        for rxn in reactions:
            l = list()
            for met in rxn.metabolites:
                l.append(get_comp(iPfal18, met.id))
            comp_dict[rxn.id] = l
        reverse_dict = dict()
        for key, value in comp_dict.items():
            try:reverse_dict[frozenset(value)].append(key)
            except:reverse_dict[frozenset(value)] = [key]
        dups = [value for key, value in reverse_dict.items() if len(value) > 1]
        if len(dups) > 0 :
            temp = dict()
            for x in dups:
                for y in x:
                    temp[y] = iPfal18.reactions.get_by_id(y).reaction
            new_dict[met_og] = temp
```

```python
# add dissipation reactions per Fritzemeier 2017 and optimize
# ATP[cytosol] + H2O[cytosol] --> ADP[cytosol] + H(+)[cytosol] + Phosphate[cytosol]
# CTP[cytosol] + H2O[cytosol] --> CDP[cytosol] + H(+)[cytosol] + Phosphate[cytosol]
# GTP[cytosol] + H2O[cytosol] --> GDP[cytosol] + H(+)[cytosol] + Phosphate[cytosol]
# UTP[cytosol] + H2O[cytosol] --> UDP[cytosol] + H(+)[cytosol] + Phosphate[cytosol]
# ITP[cytosol] + H2O[cytosol] --> IDP[cytosol] + H(+)[cytosol] + Phosphate[cytosol]
for x in ['_c','_m','_ap']:
    for y in ['atp','ctp','gtp','itp','utp']:
        met = y+x
        diphosphate_met = met.replace('t','d')
        h2o = 'h2o'+x
        h = 'h'+x
        p = 'pi'+x
        counter = 0
        for m in [h, p, h2o, diphosphate_met, met]:
            if m in [metabolite.id for metabolite in iPfal18.metabolites]:
                counter = counter
            else:
                counter = counter +1
        if counter == 0:
            with iPfal18:
                dissipation = Reaction()
                dissipation.id = 'dissipation'
                dissipation.name = 'dissipation'
                dissipation.add_metabolites({iPfal18.metabolites.get_by_id(met): -1,
                                            iPfal18.metabolites.get_by_id(h2o): -1,
                                            iPfal18.metabolites.get_by_id(diphosphate_met): +1,
                                            iPfal18.metabolites.get_by_id(h): +1,
                                            iPfal18.metabolites.get_by_id(p): +1})
                dissipation.lower_bound = 0.
                dissipation.upper_bound = 1000.

                iPfal18.add_reactions([dissipation])
                iPfal18.objective = 'dissipation'
                if iPfal18.slim_optimize()>0:
                    print(met)
                    print(iPfal18.slim_optimize())
    
```
