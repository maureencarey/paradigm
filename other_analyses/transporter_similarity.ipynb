```python
model_dict = dict()
path = "/Users/maureencarey/local_documents/work/comparative_parasite_models/paradigm/data/"
os.chdir(path)
for filename in glob.glob(os.path.join(path, '*.json')):
    key = filename.split('/')[len(filename.split('/'))-1]
    key = key[:-5]
    if key.startswith('DIY7bm_'):
        key = key[5:]
        model_dict[key] = cobra.io.load_json_model(filename)
```

```python
# these metabolites are transported INTO the cell
imported_mets_dict = dict()
for species, model in model_dict.items():
    reactants = list()
    for rxn in model.reactions:
        reactants.append([x.id for x in rxn.reactants])
    reactants = [val for sublist in reactants for val in sublist]

    transported_mets = list()
    for x in model.metabolites:
        if x.id.endswith('_e') and x.id in reactants:
            transported_mets.append(x.id[:-2])
            
    imported_mets_dict[species] = list(set(transported_mets))
```

```python
key_already_used = list()
distance_matrix_of_transporters = pd.DataFrame(index = model_dict.keys(),columns=model_dict.keys())

for species1 in model_dict.keys():
    set1 = set(imported_mets_dict[species1])
    for species2 in model_dict.keys():
        if species2 not in key_already_used:
            print(species1)
            print(species2)
            print()
            set2 = set(imported_mets_dict[species2])
            intersection_length = len(set1.intersection(set2))
            distance_matrix_of_transporters[species1].loc[species2] = intersection_length   
    key_already_used.append(species1)
distance_matrix_of_transporters.to_csv("/Users/maureencarey/local_documents/work/comparative_parasite_models/paradigm/data/transporter_similarity_before_gapfilling_july22.csv")
```
