```python
import cobra, glob, os
import pandas as pd
path = "/home/mac9jc/paradigm/models"
save_path = "/home/mac9jc/paradigm/models/excel/"
```

```python

model_dict = dict()
i = 0

for filename in glob.glob(os.path.join(path, 'gf_P*.xml')):
    key = filename.split('/')[len(filename.split('/'))-1]
    key = key[:-4]
    key = key[3:]
#    if key is not 'PconfusumCUL13' and key is not 'PneurophiliaMK1':
    model_dict[key] = cobra.io.read_sbml_model(filename)

for filename in glob.glob(os.path.join(path, 'gf_no_ortho_*.xml')):
    key = filename.split('/')[len(filename.split('/'))-1]
    key = key[:-4]
    key = key[12:]
    if key.startswith('P'):
        if key == 'PconfusumCUL13' or key == 'PneurophiliaMK1':
            model_dict[key] = cobra.io.read_sbml_model(filename)
    else:
        model_dict[key] = cobra.io.read_sbml_model(filename)
    
for key, model in model_dict.items():
    
    model_info = {'reaction ids':[r.id for r in model.reactions], 
                  'reaction formulas':[r.reaction for r in model.reactions],
                  'genes': [[g.id for g in r.genes] for r in model.reactions],
                  'upper_bounds':[r.upper_bound for r in model.reactions],
                  'lower_bounds':[r.lower_bound for r in model.reactions]}
    
    pd.DataFrame.from_dict(model_info).to_csv(save_path+key+'.csv', index = False)
    
    i = i + 1
    if i > 1: break
        
```
